```python
# Ibrar Ali, roll-no: 99, AI-Section(B), NLP_LAB-01
# N L P___L A B___T A S K S

# Task 1: Acquiring/Downloading Text
# Option B: Downloading text (requires requests library)
import requests

url = "https://www.gutenberg.org/cache/epub/1342/pg1342.txt"  # Pride and Prejudice
response = requests.get(url)
novel_text = response.text  # get only first 1000 characters
#text = response.text  # get only first 1000 characters
print(novel_text)
```

```python
# Task 2: Text Cleaning and String Functions
# 1. Lowercasing
clean_text = novel_text.lower()

# 2. Removing punctuation
import string, re
clean_text = clean_text.translate(str.maketrans("", "", string.punctuation))
clean_text = re.sub(r'[“”’]','',clean_text)

# 3. Removing numbers
clean_text = ''.join(ch for ch in clean_text if not ch.isdigit())
print("Cleaned Text:\n", clean_text)
```

```python
# Task 3: Removing Stop Words
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt_tab')

stopwords = set(stopwords.words('english'))
token_words = word_tokenize(clean_text)
filtered_text = ' '.join([word for word in token_words if word not in stopwords])
print("\nFiltered Text:",filtered_text)
```

```python
# Task 4: Word Frequency Analysis
from collections import Counter
freq_words = Counter(filtered_text.split())
print("Word Frequency Analysis: ",freq_words.most_common(10),"\n")
```
